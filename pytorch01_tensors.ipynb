## **4. Matrix Operation**

```python
f = torch.randint(size=(2,3),low=0,high=10)
g = torch.randint(size=(3,2),low=0,high=10)
```

```python
print(f)
print()
print(g)
```

```python
# matrix multiplication
print(torch.matmul(f,g))
print()
print(torch.matmul(g,f))
```

```python
v1 = torch.tensor([1,2])
v2 = torch.tensor([3,4])
# dot product
print(torch.dot(v1,v2))
print(torch.dot(v2,v1))
```

```python
# transpose
torch.transpose(f,0,1)
```

```python
h = torch.randint(size=(3,3),low=0,high=15, dtype=torch.float32)
h
```

```python
# determinant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

## **5. Comparison Operations**

```python
i = torch.randint(size=(2,3),low=0,high=15)
j = torch.randint(size=(2,3),low=0,high=15)
print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than or equal
i >= j
# less or equal
i <= j
```

# ***Tensor operations on GPU***

```python
torch.cuda.is_available()
```

```python
device = torch.device('cuda')
```

```python
# create new tensors on GPU
torch.rand((3,3),device=device)
```

```python
# Moving old tensor into GPU
a = torch.rand(2,3)
a
```

```python
b = a.to(device)
```

```python
# Testing
import torch
import time
# size of matrices
size = 10000

# random tensors on CPU
matrix_cpu1 = torch.randn(size,size)
matrix_cpu2 = torch.randn(size,size)
# Measure time on CPU
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)
cpu_time = time.time() - start_time

print(f"Time on CPU: {cpu_time: .4f} seconds")

# Move matrices to GPU
matrix_gpu1 = matrix_cpu1.to(device)
matrix_gpu2 = matrix_cpu2.to(device)

# Measure time
start_time = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)
gpu_time = time.time() - start_time

print(f"Time on GPU: {gpu_time: .4f} seconds")
```
