We were using batch gradient descent. i.e. entire dataset is going to forward pass, then loss calculation and then backward pass etc.
which has 2 major flaws:
1. **Memory ineffecient**
2. **Better Convergence**

**Solution** : *Rather than loading the entire dataset , load the batches of dataset*

```python
from sklearn.datasets import make_classification
import torch
```

# Create a Synthetic classification dataset using sklearn

```python
X, y = make_classification(
    n_samples=10,         # Number of samples
    n_features=2,         # Number of features
    n_informative=2,      # Number of informative features
    n_redundant=0,        # Number of redundant features
    n_classes=2,          # Number of classes
    random_state=42       # for reproducibility
)
```

```python
print(X.shape)
print(y.shape)
```

# Convert data to pytorch tensors

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)   # long = int64
```

# Import Dataset and DataLoader

```python
from torch.utils.data import Dataset, DataLoader
```

# Create a custom class

```python
class CustomDataSet(Dataset):
  def __init__(self, features, labels):
    self.features = features         # features is X
    self.labels = labels             # labels is y

  def __len__(self):
    return self.features.shape[0]       # or len(self.features)

  def __getitem__(self,index):
    return self.features[index], self.labels[index]
```

# Create an instance of CustomDataSet

```python
dataset = CustomDataSet(X,y)
```

```python
len(dataset)
```

```python
dataset[8]
```

# Create an instance of DataLoader

```python
# mention batch_size and shuffle parameter
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
# batches will form of 2 and it would be randomly arranged

for batch_features, batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)
  print("-"*50)
```
